# Imports



*   TensorFlow
*   Numpy
*   Keras





```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
print(tf.__version__)
```


# Define and compile a simplest neural network 

With 1 layer and 1 neuron and with just 1 input shape value. We will use Sequential class and Dense layer.


```python
model = tf.keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
```

Now we compile the neural network with optimizer and loss function i,e y = 2x - 1. Loss function compares the result with the known correct answers and measures how badly it was done.
Then passes on to optimizer which makes another prediction but with performance better than the previous prediction.


```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

# Providing the data

Now we feed the data, In this case we are using 6X and 6Y values. With relationship between these as y = 2x -1.
we use numpy to declare the values.

```python
xs = np.array([-1.0, 0, 1, 2, 3, 4]);
ys = np.array([-3, -1, 1, 3, 5, 7]);
```

#Training the Neural network
We are asking the model to fit the xs to the ys and we do the 500 epocs. It goes through every iteration and makes a guess and measure how bad it has performed and does better next time.

```python
model.fit(xs,ys,epochs=500)
```

Now we have a trained model and we use it to predict the y unknown for the x given.

```python
print(model.predict([10.0]))
```
